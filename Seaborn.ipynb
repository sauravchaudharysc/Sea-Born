# Introduction to Seaborn

Seaborn is an amazing visualization library for statistical graphics plotting in Python.It provides default styles and color palettes to make statistical plots more attractice.It is buils on the top of matplotlib library and closely integrated to the data structures from pandas.Seaborn aims to make visualisation the central part of exploring and understanding data.It provides dataset-oriented APIs,so that we can switch between different visual representations for same variables for better understanding of dataset.

```python
#Import Seaborn
import seaborn as sns
import numpy as np
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#A normal count plot to count the frequency of data. So there is no Y-axis
sns.countplot(x='sex',data=tips)
```

```python
#This indicates most of the people come for dinner
sns.countplot(x='time',data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
#It indicates the maximum and minimum bill on particular day
#On Thursday average minimum bill is 7 and maximum is 20
```

```python
#To categorize the days where smoker are maximum
sns.boxplot(x="day",y="total_bill",data=tips,hue='smoker')
```

```python
#To categorize the days on gender
sns.boxplot(x="day",y="total_bill",data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

## Distribution Plot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
#We can observer that as the bill is increasing then tip is increasing.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
#We can observer that as the bill is increasing then tip is increasing.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
#We can observer that as the bill is increasing then tip is increasing.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
#We can observer that as the bill is increasing then tip is increasing.
```

```python
sns.pairplot(tips)
```

```python
#to differentiate on the basis of gender
sns.pairplot(tips,hue='sex')
```

```python
tips_corr=tips.corr()
```

```python
tips_corr
```

```python
sns.heatmap(tips_corr,annot=True)
```
